## 8. Detecting Outliers

An outlier is a point or set of points that are different from other points.

It's often a good idea to remove outliers as they're one of the primary reasons for resulting in a less accurate model

```python
print(df.columns)
```

```python
sns.boxplot(x = df['Rating'])
```

```python
sns.boxplot(x = df['Price range'])
```

```python
sns.boxplot(x = df['Average Cost for two'])
```

```python
def RemoveOutliers(df, Column, lowerThreshold, upperThreshold):

  sns.boxplot(x = df[Column])
  plt.title(f'Original Box Plot of {Column}')
  plt.show()

  df = df[(df[Column] >= lowerThreshold) & (df[Column] <= upperThreshold)]

  sns.boxplot(x = df[Column])
  plt.title(f'After Removing Outliers Box Plot of {Column}')
  plt.show()

  return df

RemoveOutliers(df, 'Rating', 1, 5)
```

## 9. Plot different features against one another

### Histogram

Histogram refers to the frequency of occurence of variables in an interval.

Here, nlargest(n) describes number of bars

```python
df.columns
```

```python
df['Rating'].value_counts().nlargest(40).plot(kind = 'bar', figsize = (10, 5))
plt.title("Price Range vs Ratings")
plt.xlabel('Rating')
plt.ylabel('Price Range')
```

```python
plt.figure(figsize = (10, 6))
sns.histplot([df['Rating']])
```
